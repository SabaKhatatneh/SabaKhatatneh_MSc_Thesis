```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
from scipy.stats import spearmanr
from scipy.stats import pearsonr
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#Drug signatures file
signatures = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/Drugs/results/consensus_signature_drugs.csv',
                        sep=',', header=0, index_col=0)
signatures
```

```python
#Meta (drugs MOA) file
meta = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/Drugs/results/drugs_meta.csv',
                  sep=',', header=0, index_col=0)
meta
```

```python
#calculating spearman correlation of the drug signatures 
signature_similarity = spearmanr(signatures.T)
```

```python
# correlation values
signature_similarity[0]
# p values
signature_similarity[1]
```

```python
signature_similarity = signature_similarity[0]
```

```python
#creating a dataframe of the spearman correlation values.
signature_similarity = pd.DataFrame(signature_similarity, index=signatures.index, columns=signatures.index)
```

```python
signature_similarity
```

### Create the shared MOA matrix

```python
#multiply the meta (MOA) matrix to its inverse.
meta_similarity = np.dot(meta.values, meta.T.values)
```

```python
meta_similarity = pd.DataFrame(meta_similarity, index=meta.index, columns=meta.index)
```

```python
meta_similarity
```

```python
#equalize the non-zero values to 1.
meta_similarity = (meta_similarity > 0) * 1
```

```python
meta_similarity
```

###### Mann whitney U test between drugs with same MOA and drugs without

```python
### select lower triangular matrix, without diagonal (self correlations)
fil = np.tril_indices_from(signature_similarity, k=-1)
### make vectors from the lower triangular matrix
signature_similarity_tril = signature_similarity.values[fil]
meta_similarity_tril = meta_similarity.values[fil]
### select instances where the MoA is same (meta_similarity==1)
fil = meta_similarity_tril == 1
### shared MoA correlatons: A[fil]
### different MoA correlation A[~fil]
mannwhitneyu(signature_similarity_tril[fil], signature_similarity_tril[~fil])
```

```python
fil = np.tril_indices_from(meta_similarity.values, k=-1)
```

```python
meta_similarity_lower_triangular = meta_similarity.values[fil]
```

```python
signature_similarity_lower_triangular = signature_similarity.values[fil]
```

```python
y_tr = meta_similarity_lower_triangular
y_pr = signature_similarity_lower_triangular
```

```python
fpr, tpr, thr = roc_curve(y_true=y_tr, y_score=y_pr)
```

```python
auc = roc_auc_score(y_true=y_tr, y_score=y_pr)
plt.plot(fpr, tpr, label = "AUC score = %s" %auc, color = "green" )
# linestyle = "--"
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Prediction Performance")
plt.legend(loc="lower right")
plt.savefig('drug signature')
plt.show()
```
