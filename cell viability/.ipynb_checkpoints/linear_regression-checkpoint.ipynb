```python
import pandas as pd
import numpy as np
```

### Open files needed 
#### 1. disease id - concept id matrix
#### 2. Drugs signatures file
#### 3. Disease Signatures file
#### 4. Achilles dataset

```python
# open the file contains concept ids and BRD ids.
drug_concept = pd.read_csv('../data/results/BRD_concept_matrix.csv', sep=',', header=0, index_col=0)
drug_concept
```

```python
# open the file contains concept ids and disease ids.
disease_concept = pd.read_csv('../data/results/Concept_dz.csv', sep=',', header=0, index_col=0)
disease_concept
```

```python
# Drugs signatures file
drugs_signature = pd.read_csv('../data/Drugs/consensus_signature_drugs.csv', sep=',', header=0, index_col=0)
drugs_signature
```

```python
# disease signatures file
disease_signature = pd.read_csv('../data/Diseases/merge_and_overlap.csv', sep=',', header=0, index_col=0)
disease_signature = disease_signature.T
```

```python
disease_signature
```

```python
# achilles dataset file
model = pd.read_csv('../data/linear model/achilles.csv', sep=',', header=0, index_col=0)
model
```

```python
# remove intercept row
model_1 = pd.DataFrame()
model_1 ['genes'] = model['pr_gene_symbol']
model_1 ['coeff'] = model['coefficient']
model_1 = model_1.drop('INTERCEPT', axis=0)
```

```python
model_1
```

```python
# set genes column as index
model_1 = model_1.set_index(['genes'])
model_1
```

#### Calculate cell viability using linear regression equation

```python
# find the common genes between drugs signatures and achilles model
common_genes = list(set(drugs_signature.columns) & set(model_1.index))
```

```python
# find the common genes between diseases signatures and achilles model
dz_common_genes = list(set(disease_signature.columns) & set(model_1.index))
```

```python
# find the cell viability values using drugs signatures and achilles model (using lineaer regression equation)
drug_cv_values = []
for row in drugs_signature.index:
    cv = 0
    for gene in drugs_signature.columns:
        cv += model_1.loc[gene, 'coeff'] * drugs_signature.loc[row, gene]
    drug_cv_values.append(cv)
print(drugs_signature.loc[row, gene])
```

```python
# find the cell viability values using diseases signatures and achilles model (using lineaer regression equation)
dz_cv_values = []
for row in disease_signature.index:
    a = (disease_signature.loc[row] * model_1['coeff']).sum()
    dz_cv_values.append(a)
```

```python
## another method for calculating cell viability (in case of matching dimensions of the matrices)
# create a dataframe contains drugs BRD ids and cell viability values.
drugs_cv_df = drugs_signature @ model_1
```

```python
drugs_cv_df
```

```python
# create a dataframe contains disease ids and cell viability values
diseases_cv_df = pd.DataFrame()
diseases_cv_df['disease Id'] = disease_signature.index
diseases_cv_df['cell viability'] = dz_cv_values
```

```python
diseases_cv_df
```

#### Filter cancer diseases only, using the cancer parent concept id (C0006826) Malignant neoplastic disease

```python
disease_concept = disease_concept.T
```

```python
disease_concept
```

```python
# using the parent concept id of cancer diseases to filter out cancer disease Ids.
disease_concept['C0006826']
```

```python
# create a series which contains cancer disease ids only!  
cancer_ser = pd.Series(data=disease_concept['C0006826'].values, index=disease_concept.index)
```

```python
# filter out the disease ids that are not related to cancer parent concept id 'C0006826'. (values are 0)
fil = cancer_ser == 1
cancer_ser = cancer_ser[fil]
```

```python
# change the series to a dataframe, and rename the column
cancer_df = cancer_ser.to_frame()
cancer_df = cancer_df.rename(columns={0: 'C0006826'})
```

```python
cancer_df
```

```python
disease_concept.insert(0, 'disease id', disease_concept.index)
disease_concept
```

### Create cancer disease, non-cancer disease dataframes
#### using the Cancer df created from the parent concept id of cancers

#### Cancer diseases dataframe

```python
# create a dataframe contains cancer disease ids
cancer_concept = disease_concept[disease_concept['disease id'].isin(cancer_ser.index)]
cancer_concept = cancer_concept.set_index('disease id')
cancer_concept
```

```python
# eliminate concept ids that are not related to the cancer disease ids.
for i in cancer_concept.columns:
    if cancer_concept[i].sum() == 0:
        cancer_concept= cancer_concept.drop(columns = [i])
```

```python
cancer_concept
```

#### non-Cancer diseases dataframe

```python
# create non-cancer diseases matrix
non_cancer_concept = disease_concept[~disease_concept['disease id'].isin(cancer_ser.index)]
non_cancer_concept = non_cancer_concept.set_index('disease id')
non_cancer_concept
```

```python
# eliminate concept ids that are related to the cancer disease ids.
for i in non_cancer_concept.columns:
    if non_cancer_concept[i].sum() == 0:
        non_cancer_concept= non_cancer_concept.drop(columns = [i])
```

```python
non_cancer_concept
```

### Creating anti-cancer drugs, and non-cancer drugs dataframes

```python
# finding the common cancer concept ids.
common_dr_con = list(set(drug_concept.columns) & set(cancer_concept.T.index))
```

```python
drug_concept2= drug_concept.T
drug_concept2.insert(0, 'concepts', drug_concept2.index)
drug_concept2
```

#### Anti-Cancer drugs dataframe

```python
# filter out the concept ids that are not in the common list of cancer concept ids from the dataframe.
cancer_drugs = drug_concept2[drug_concept2['concepts'].isin(common_dr_con)]
cancer_drugs = cancer_drugs.drop(columns=['concepts'])
```

```python
cancer_drugs
```

```python
# eliminate BRD ids that are not correlated to the cancer concept ids.
for i in cancer_drugs.columns:
    if cancer_drugs[i].sum() == 0:
        cancer_drugs= cancer_drugs.drop(columns = [i])
```

```python
cancer_drugs
```

#### non-Cancer drugs dataframe

```python
# create a non-cancer drugs matrix
non_cancer_drugs = drug_concept2[~drug_concept2['concepts'].isin(common_dr_con)]
non_cancer_drugs = non_cancer_drugs.drop(columns=['concepts'])
```

```python
non_cancer_drugs
```

```python
# eliminate BRD ids that are not correlated to the non-cancer concept ids.
for i in non_cancer_drugs.columns:
    if non_cancer_drugs[i].sum() == 0:
        non_cancer_drugs= non_cancer_drugs.drop(columns = [i])
```

```python
non_cancer_drugs
```
