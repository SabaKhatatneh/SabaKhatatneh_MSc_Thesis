```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cancer diseases ids vs cancer related concept ids matrix 
cancer_concept = pd.read_csv('../data/linear model/results/cancer_diseases.csv', sep=',', header=0, index_col=0)
cancer_concept
```

```python
# non-cancer diseases ids vs concepts ids 'not cancer related' matrix
non_cancer_concept = pd.read_csv('../data/linear model/results/non_cancer_diseases.csv', sep=',', header=0, index_col=0)
non_cancer_concept
```

```python
# concept ids vs non-cancer drug ids matrix
non_cancer_drugs = pd.read_csv('../data/linear model/results/non_cancer_drugs.csv', sep=',', header=0, index_col=0)
non_cancer_drugs
```

```python
# cancer concept ids vs cancer drugs matrix
cancer_drugs = pd.read_csv('../data/linear model/results/cancer_drugs.csv', sep=',', header=0, index_col=0)
cancer_drugs
```

```python
# drugs cell viability values
drugs_cv_df = pd.read_csv('../data/linear model/results/drugs_cv_df.csv', sep=',', header=0, index_col=0)
drugs_cv_df
```

```python
# diseases cell viability values
diseases_cv_df = pd.read_csv('../data/linear model/results/diseases_cv_df.csv', sep=',', header=0, index_col=0)
diseases_cv_df = diseases_cv_df.rename(columns={'disease Id': 'disease_id'})
diseases_cv_df
```

```python
# adjusting the disease ids name
diseases_cv_df.disease_id = [s.replace("-", ":") for s in diseases_cv_df.disease_id]
diseases_cv_df = diseases_cv_df.set_index('disease_id')
diseases_cv_df
```

### Create cell viability tables

##### Diseases cell viability tables

```python
# finding the common disease-ids between diseases-cell viability table and 'cancer disease-ids vs concept-ids' matrix
common_dz_cn = list(set(cancer_concept.T.columns) & set(diseases_cv_df.index))
```

```python
# create a table contains cancer disease-ids cell viability values.
cn_dz_cv_ =  diseases_cv_df[diseases_cv_df.index.isin(common_dz_cn)]
cn_dz_cv_ = cn_dz_cv_.reset_index()
cn_dz_cv_.head()
```

```python
# create a table contains non-cancer disease-ids cell viability values.
noncn_dz_cv_ =  diseases_cv_df[~diseases_cv_df.index.isin(common_dz_cn)]
noncn_dz_cv_ = noncn_dz_cv_.reset_index()
noncn_dz_cv_
```

##### Drugs cell viability tables

```python
# finding the common drugs broad-ids between drug-cell viability table and 'cancer drugs broad-ids vs concept-ids' matrix
common_BRD_cn = list(set(cancer_drugs.columns) & set(drugs_cv_df.index))
```

```python
# create a table contains cancer drugs Broad-ids cell viability values.
cn_drugs_cv_ =  drugs_cv_df[drugs_cv_df.index.isin(common_BRD_cn)]
cn_drugs_cv_ = cn_drugs_cv_.reset_index()
cn_drugs_cv_
```

```python
# create a table contains non-cancer drugs Broad-ids cell viability values.
noncn_drugs_cv_ =  drugs_cv_df[~drugs_cv_df.index.isin(common_BRD_cn)]
noncn_drugs_cv_ = noncn_drugs_cv_.reset_index()
noncn_drugs_cv_
```

```python
# Boxplot of Cancer diseases cell viability values and Non-Cancer diseases cell viability values
fig = plt.figure()
box = plt.boxplot ([cn_dz_cv_['cell viability'], noncn_dz_cv_['cell viability']],whis=[5, 9], 
                   showfliers= True, widths=0.4, patch_artist=True, )
plt.title('Diseases')

colors = ['rosybrown', 'darkseagreen']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
        
fliers = box['fliers']
for fly in fliers:
    fdata = fly.get_data()
    fly.set_data([fdata[0][0],fdata[0][-1]],[fdata[1][0],fdata[1][-1]])

plt.ylim([-2,2.5])
    
plt.ylabel('cell viability')
plt.xticks([1, 2], ['cancer diseases', 'non cancer diseases'])
```

```python
# Boxplot of Cancer drugs cell viability values and Non-Cancer drugs cell viability values
fig = plt.figure()
box = plt.boxplot ([cn_drugs_cv_['coeff'], noncn_drugs_cv_['coeff']], whis=[5, 9], 
                   showfliers= True, widths=.4, patch_artist=True)

colors = ['lightgreen' , 'tan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    
fliers = box['fliers']
for fly in fliers:
    fdata = fly.get_data()
    fly.set_data([fdata[0][0],fdata[0][-1]],[fdata[1][0],fdata[1][-1]])

plt.ylim([-4,1])
plt.title('Drugs')
plt.ylabel('cell viability')
plt.xticks([1, 2], ['cancer drugs', 'non cancer drugs'])
```

### Using statistical test to compare between cancer and non-cancer diseases groups and drugs groups

```python
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
```

### Mann–Whitney U test

```python
# compare between diseases
mannwhitneyu(noncn_dz_cv_['cell viability'],cn_dz_cv_['cell viability'])
```

```python
# compare between drugs
mannwhitneyu(noncn_drugs_cv_['coeff'],cn_drugs_cv_['coeff'])
```

```python
# compare between drugs
mannwhitneyu(cn_dz_cv_['cell viability'],cn_drugs_cv_['coeff'])
```

### T-test (for independent samples)

```python
# compare between diseases
ttest_ind(noncn_dz_cv_['cell viability'],cn_dz_cv_['cell viability'])
```

```python
# compare between drugs
ttest_ind(noncn_drugs_cv_['coeff'],cn_drugs_cv_['coeff'])
```
