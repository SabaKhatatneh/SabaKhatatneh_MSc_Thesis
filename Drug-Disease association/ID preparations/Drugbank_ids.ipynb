##### Drug bank vocabulary

```python
drugbank = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/drugbank vocabulary.csv', sep=',', header=0, index_col=0)
drugbank
```

```python
fil = drugbank['Standard InChI Key'].isna()
drugbank = drugbank[~fil] # ~ not fil (Inverts all the bits)
```

```python
# split the inchi key 
# [0] and remove the rest of the number after (-)
# choose the first 16 numbers
drugbank['inchi_key_prefix'] = drugbank['Standard InChI Key'].apply(lambda x: x.split('-')[0])
```

```python
# store multiple items in a variable (inchi keys in common-inchi)
common_inchi = set(Drugs['common_inchi']) & set(drugbank['inchi_key_prefix'])
```
