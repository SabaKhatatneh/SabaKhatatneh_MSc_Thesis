```python
import pandas as pd
import numpy as np
```

```python
# open matched Drug BRD ID - DrugBank ID file
brd_db = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/drugIds.csv', sep=',', header=0, index_col=0)
brd_db = brd_db[['BRD-ID', 'DB-ID']]
brd_db
```

```python
# add an indicatior column
brd_db['Indicator'] = 1
brd_db.head()
```

```python
# open repoDB Concept ID-DrugBank ID file
db_conc = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/repoDB_match.csv', sep=',', header=0, index_col=0)
db_conc.columns = ['concept_id', 'DB-ID']
db_conc
```

```python
# add an indicator column
db_conc['Indicator'] = 1
db_conc.head()
```

```python
# open matched disease DOID-concept ID file
conc_doi = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/DOIDs_CONCEPTs.csv', sep=',', header=0, index_col=0)
conc_doi.columns = ['doid', 'concept_id']
conc_doi
```

```python
# add an indicator column
conc_doi['Indicator'] = 1
conc_doi.head()
```

### Creating matrices of the Identifiers 

```python
# creating the disease IDs matrix
conc_doi_ma = pd.pivot_table(conc_doi, index='concept_id', columns='doid', values='Indicator')
conc_doi_ma[conc_doi_ma.isna()] = 0
conc_doi_ma = conc_doi_ma.astype(int)
conc_doi_ma
```

```python
# create the drug IDs matrix
brd_db = pd.pivot_table(brd_db, index='BRD-ID', columns='DB-ID', values='Indicator')
brd_db[brd_db.isna()] = 0
brd_db = brd_db.astype(int)
```

```python
brd_db.head()
```

```python
# create a matrix of the repoDB derived concept ID and DrugBank ID
# Indication Matrix, where 1 is between the drug ID and the corresponding indicated disease ID.
db_conc = pd.pivot_table(db_conc, index='DB-ID', columns='concept_id', values='Indicator')
db_conc[db_conc.isna()] = 0
db_conc = db_conc.astype(int)
```

```python
db_conc
```

### create an Indication matrix (drugs vs diseases)

```python
# Using drug IDs from LINCS database matched with repoDB drug IDs
```

```python
# find the common drugs between LINCS and repoDB using DrugBank IDs
common_db = list(set(brd_db.columns) & set(db_conc.index))
```

```python
# multiply Drug IDs matrix (contains the common_db only) with repoDB matrix (containing the common_db only)
# Creating indication matrix from the common drugs (common_db)
brd_conc = brd_db[common_db] @ db_conc.loc[common_db]
```

```python
brd_conc
```

```python
# equalizing values >0 to 1, representing drug and disease indication
brd_conc[brd_conc>0] = 1
```

```python
# brd_conc.to_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/brd_conc.csv')
```

```python
# Using disease IDs from CREEDS database matched with repoDB disease IDs
```

```python
# find the common diseases between CREEDS and brd_conc matrix using Concept IDs
### concept IDs in brd_conc are derived from repoDB
common_conc = list(set(brd_conc.columns) & set(conc_doi_ma.index))
```

```python
# multiply Disease IDs matrix (contains the common_conc only) with brd_conc matrix (containing the common_conc only)
# Creating indication matrix from the common disease (common_conc)
conc_doi_ma = brd_conc[common_conc] @ conc_doi_ma.loc[common_conc]
```

```python
# equalizing values >0 to 1, representing drug and disease indication
conc_doi_ma[conc_doi_ma>0] = 1
```

```python
# final indication matrix with DOIDs and BRD IDs
conc_doi_ma
```
