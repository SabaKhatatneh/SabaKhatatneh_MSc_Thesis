```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

#### prepare the tables for plotting a roc curve

```python
# open all diseases concept id vs disease ids
dz_con = pd.read_csv('../data/results/ancestors_offsprings.csv', sep=',', header=0, index_col=0)
dz_con= dz_con[['dz_id', 'concept', 'type']]
dz_con= dz_con.reset_index(drop=True)
dz_con
```

```python
# filter self diseases
self_df = dz_con.loc[dz_con['type'] == 'self']
self_df
```

```python
# filter ancestors diseases
ancestors_df = dz_con.loc[dz_con['type'] == 'ancestor']
ancestors_df
```

```python
# add indicator column to the table and adding an indicator column
self_anc_df = pd.concat([self_df, ancestors_df])
self_anc_df = self_anc_df.reset_index(drop=True)
self_anc_df = self_anc_df[['dz_id','concept', 'type']]
self_anc_df['Indicator']= 1
self_anc_df
```

```python
# reshape the table into a matrix 
self_anc_df = pd.pivot_table(self_anc_df, index='concept', columns='dz_id', values='Indicator')
self_anc_df[self_anc_df.isna()] = 0
self_anc_df = self_anc_df.astype(int)
self_anc_df
```

```python
# open the indication matrix
brd_conc = pd.read_csv('../data/results/brd_conc.csv', sep=',', header=0, index_col=0)
brd_conc
```

```python
# find the common concept Ids between the matrices
common_db = list(set(brd_conc.columns) & set(self_anc_df.index))
```

```python
# multiplying both matrices with the common diseases to get the common indication matrix
self_anc_brd_dz = brd_conc[common_db] @ self_anc_df.loc[common_db]
```

```python
# correlation matrix (disease ID vs Broad ID)
similarity = pd.read_csv('../data/results/brd_dz_corr.csv', sep=',', header=0, index_col=0)
similarity
```

```python
# adjust the column names
good_columns = []
for c in self_anc_brd_dz.columns:
    good_columns.append(c.replace(':','-'))
```

```python
self_anc_brd_dz.columns = good_columns
```

#### adjust the matrices to be same dimensions with the common drug and disease IDs

```python
# find the common drugs between the indication and correlation matrices
common_brd = list(set(self_anc_brd_dz.index) & set(similarity.index))
```

```python
# find the common diseases between the indication and the correlation matrices
common_dz = list(set(self_anc_brd_dz.columns) & set(similarity.columns))
```

```python
self_anc_brd_dz =  self_anc_brd_dz.loc[common_brd,common_dz]
```

```python
similarity =  similarity.loc[common_brd,common_dz]
```

#### Create roc curve from the matrices with same dimensions

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# equalize any value higher than 1 to 1
self_anc_brd_dz[self_anc_brd_dz>0] = 1
```

```python
y_tr = self_anc_brd_dz.values.reshape((1, -1))[0] # ground truth value = indication matrix
y_pr = similarity.values.reshape((1,-1))[0] # predicted values = correlation matrix
```

```python
fpr, tpr, thr = roc_curve(y_true=y_tr, y_score=y_pr)
```

```python
# plot the roc curve with the AUC score
auc = roc_auc_score(y_true=y_tr, y_score=y_pr)
plt.plot(fpr, tpr, label ="Our prediciton. AUC = %s" %auc, color = "green" )
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Self-Ancestors Prediction Performance")
plt.legend(loc="lower right")
plt.plot([0,1],[0,1], linestyle = "--", color = "red")
plt.show()
```
