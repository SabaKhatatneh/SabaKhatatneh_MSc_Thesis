```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Create drug-disease signatures correlation matrix

```python
# open disease signatures file
disease = pd.read_csv('../data/Diseases/merge_and_overlap.csv', sep=',', header=0, index_col=0)
```

```python
# open drug signatures file
drug = pd.read_csv('../Drugs/results/consensus_signature_drugs.csv', sep=',', header=0, index_col=0)
```

##### use the common genes between disease signatures and drug signatures

```python
# find the common genes between drugs and diseases signatures
genes = list(set(drug.columns) & set(disease.index))
```

```python
disease = disease.loc[genes]
drug = drug[genes]
```

```python
# use the gene with total null values < 300
fil = disease.isna().sum() < 300
disease = disease[disease.columns[fil]]
```

```python
drug = drug.T
```

```python
# concatenate both matrices
all_data = pd.concat([drug, disease], axis=1)
```

```python
all_data
```

```python
# calculate the correlation between all columns
similarity = all_data.corr()
```

```python
similarity
```

```python
# restructure the matrix where drugs Ids are the rows and disease Ids are the columns
similarity = similarity.loc[drug.columns, disease.columns]
```

```python
### The final correlation matrix between drug and disease signatures
similarity
```

```python
# open CREEDS disease signatures file
creeds = pd.read_csv('../data/Diseases/disease_signatures-v1.0.csv',
                    sep=',', header=0, index_col=0)
creeds['dz_Ids']= creeds.index
creeds
```

```python
creeds
```

```python
# create a table from CREEDS file, containing dz IDs and DOID
## add an indicator column
dz_doid = pd.DataFrame()
dz_doid['dz_id'] = creeds['dz_Ids']
dz_doid['do_id'] = creeds['do_id']
dz_doid['Indicator'] = 1
dz_doid = dz_doid.reset_index(drop=True)
dz_doid
```

```python
# reshape the table into a matrix
## 1 indicates dz Id related to the corresponding DOID
dz_doid_ma = pd.pivot_table(dz_doid, index='do_id', columns='dz_id', values='Indicator')
dz_doid_ma[dz_doid_ma.isna()] = 0
dz_doid_ma = dz_doid_ma.astype(int)
dz_doid_ma
```

```python
# open the indication matrix (BRD ID as drug Ids vs DOID as disease Ids)
brd_doid = pd.read_csv('../data/results/BRD_doid_matrix.csv', sep=',', header=0, index_col=0)
```

```python
brd_doid
```

#### the indication matrix and CREEDS disease identifiers matrix

```python
# find the common DOID
common_doid = list(set(brd_doid.columns) & set(dz_doid_ma.index))
```

```python
# multiply both matrices to get an indication matrix with the dz Ids vs BRD Ids
brd_diz = brd_doid[common_doid] @ dz_doid_ma.loc[common_doid]
```

```python
# equalizing all values more than zero to 1
brd_diz[brd_diz>0] = 1
```

```python
# adjusting the column names
good_columns = []
for c in brd_diz.columns:
    good_columns.append(c.replace(':','-'))
brd_diz.columns = good_columns
```

```python
# the final indication matrix
brd_diz
```

### Creating a Roc curve of correlation matrix vs Indication matrix

#### match the dimensions of the correlation and indication matrices

```python
# find the common drug Broad Ids
common_brd = list(set(brd_diz.index) & set(similarity.index))
```

```python
# find the common disease dz Ids
common_dz = list(set(brd_diz.columns) & set(similarity.columns))
```

```python
# use the common diseases and drugs in the matrices
brd_diz =  brd_diz.loc[common_brd,common_dz]
similarity =  similarity.loc[common_brd,common_dz]
```

```python
brd_diz
```

```python
similarity
```

### Create a roc curve from the indication matrix and the correlation matrix

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
y_tr = brd_diz.values.reshape((1, -1))[0] # ground truth values = indication matrix
y_pr = similarity.values.reshape((1,-1))[0] # predicted values = correlation matrix
```

```python
fpr, tpr, thr = roc_curve(y_true=y_tr, y_score=y_pr)
```

```python
# plot the roc curve with the AUC score
auc = roc_auc_score(y_true=y_tr, y_score=y_pr)
plt.plot(fpr, tpr, label ="Our prediciton. AUC = %s" %auc, color = "green" )
# linestyle = "--"
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Disease-Drug similarity Prediction Performance")
plt.legend(loc="lower right")
plt.plot([0,1],[0,1], linestyle = "--", color = "red")
plt.show()
```
