```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.stats import mannwhitneyu
```

```python
# read disease identifiers table
table = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/Drugs/results/All_DOIDs_CONCEPTs_diseases.csv', sep=',', header=0, index_col=0)
table
```

```python
# create a dictionary of concept Ids and disease names
keys = table['concepts']
values = table['diseases']
table_dic = {keys[i]: values[i] for i in range(len(keys))}
```

```python
# read the file of disease-concept ids 
dz_con = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/ancestors_offsprings.csv', sep=',', header=0, index_col=0)
dz_con= dz_con[['dz_id', 'concept', 'type']]
dz_con= dz_con.reset_index(drop=True)
dz_con
```

```python
# Indication matrix (broad id of drugs vs concept ids of diseases)
brd_conc = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/brd_conc.csv', sep=',', header=0, index_col=0)
brd_conc
```

```python
# open the correlation matrix (broad id of drugs vs disease ids of diseases)
similarity = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/results/brd_dz_corr.csv', sep=',', header=0, index_col=0)
similarity 
```

```python
# adjusting the colum names (disease ids)
good_columns = []
for c in similarity.columns:
    good_columns.append(c.replace('-',':'))
```

```python
similarity.columns = good_columns
```

```python
# filtering the self concept ids only
self_df = dz_con.loc[dz_con['type'] == 'self']
self_df = self_df.reset_index(drop=True)
self_df = self_df[['dz_id','concept']]
self_df
```

```python
#filtering offspring concept ids only
offspring_df = dz_con.loc[dz_con['type'] == 'offspring']
offspring_df
```

```python
# concatenate both tables together (offsprings, self)
self_offsp_df = pd.concat([self_df, offspring_df])
self_offsp_df = self_offsp_df.reset_index(drop=True)
self_offsp_df = self_offsp_df[['dz_id','concept']]
self_offsp_df
```

```python
concept_unique = []
concept_unique = self_offsp_df['concept'].unique()
```

```python
## Applying Man-whitney U test between diseases in the corr matrix and indication matrix
## and record the resulted P-value from each comparison in the table and the AUC scores

df_app = pd.DataFrame()
df_nonapp = pd.DataFrame()
p_value_list = []
disease = []

AUC_scores = []
conc_id_list = []
non_match = []
conc_AUC_table = pd.DataFrame ()
fig, (ax1,ax2) = plt.subplots(nrows=2, sharex=True, sharey=True)
plt.ion()
# dz_AUC['BRD'] = brd_conc.index
for i in concept_unique:
    self_off_conc_df = self_offsp_df.loc[self_offsp_df['concept'] == i]
    self_off_conc_df['Indicator']= 1
    self_off_conc_df = pd.pivot_table(self_off_conc_df, index='concept', columns='dz_id', values='Indicator')
    self_off_conc_df[self_off_conc_df.isna()] = 0
    self_off_conc_df = self_off_conc_df.astype(int)
    common_db = list(set(brd_conc.columns) & set(self_off_conc_df.index))
    if len(common_db)>0:
        conc_df_brd_dz = brd_conc[common_db] @ self_off_conc_df.loc[common_db]
        common_brd = list(set(conc_df_brd_dz.index) & set(similarity.index))
        if len(common_brd)>0:
            common_dz = list(set(conc_df_brd_dz.columns) & set(similarity.columns))
            if len(common_dz)>0:
                p_table = pd.DataFrame()
                conc_ed =  conc_df_brd_dz.loc[common_brd,common_dz]
                sim_df =  similarity.loc[common_brd,common_dz]
                conc_ed[conc_ed>0] = 1
                y_tr = conc_ed.values.reshape((1, -1))[0]
                y_pr = sim_df .values.reshape((1,-1))[0]
                p_table['pred'] = y_tr
                p_table['corr'] = y_pr
                df_app = p_table.loc[p_table['pred'] == 1]
                df_nonapp = p_table.loc[p_table['pred'] == 0]
                stat, p = mannwhitneyu(df_app['corr'], df_nonapp['corr'])
                if y_tr.sum()>0:
                    p_value_list.append(p)
                    disease.append(table_dic[i])
                    AUC = roc_auc_score(y_true=y_tr, y_score=y_pr)
                    AUC_scores.append(AUC)
                    conc_id_list.append(i) 
conc_AUC_table['conc_ids'] = conc_id_list
conc_AUC_table['AUC'] = AUC_scores
conc_AUC_table['p_value'] = p_value_list
conc_AUC_table['diseases'] = disease
```
