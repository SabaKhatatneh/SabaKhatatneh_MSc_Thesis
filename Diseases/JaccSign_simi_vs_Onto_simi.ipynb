```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.stats import spearmanr
from scipy.stats import pearsonr
```

```python
#open Jaccard signature similarity file
sig_sim = pd.read_csv('../results/general_similiraty.csv'
                      , sep=',', header=0, index_col=None)
#open the doSim similarity file
ont_sim = pd.read_csv('../results/doidstable.csv'
                      , sep=',', header=0, index_col=None)
```

```python
sig_sim.head()
```

```python
ont_sim.head()
```

```python
#open CREEDS dataset file
meta = pd.read_csv('../data/disease_signatures-v1.0.csv'
                   , sep=',', header=0, index_col=None)
```

```python
meta.head()
```

### Reshaping and preparing the dataframes 

```python
#Jaccard similarity table reshaping
sig_sim = pd.melt(sig_sim, id_vars='Unnamed: 0', value_vars=sig_sim.columns[1:])
sig_sim.columns = ['DZ1', 'DZ2', 'SigSim']
```

```python
sig_sim
```

```python
# Removing self correlation values
fil = sig_sim != 2
sig_sim  = sig_sim [fil]
```

```python
# removing null values
fil = sig_sim.isna().sum(1) == 0
sig_sim = sig_sim[fil]
```

```python
# dividing the correlation values on 2 (the general similarity file is a result of summed similarities)
sig_sim['SigSim']= sig_sim['SigSim'].div(2)
```

```python
sig_sim.head()
```

### Here we will do the same for doSim similarity dataframe

```python
#doSim similarity table reshaping
ont_sim = pd.melt(ont_sim, id_vars='Unnamed: 0', value_vars=ont_sim.columns[1:])
ont_sim.columns = ['DO1', 'DO2', 'OntSim']
```

```python
# Removing self correlation values
fil = ont_sim != 1
ont_sim  = ont_sim [fil]
```

```python
# removing null values

fil = ont_sim.isna().sum(1) == 0
ont_sim = ont_sim[fil]
```

```python
ont_sim.head()
```

### Creating a merged table of sig_sim (Jaccard) and ont_sim (doSim)

```python
meta = meta[['id', 'do_id']]
```

```python
meta
```

```python
meta.index = meta['id']
meta = meta['do_id']
```

```python
# meta [ sig_sim variables' values]
sig_sim['DO1'] = meta[sig_sim['DZ1'].values].values
sig_sim['DO2'] = meta[sig_sim['DZ2'].values].values
```

```python
sig_sim
```

```python
# remove null data
fil = sig_sim.isna().sum(1) == 0
sig_sim = sig_sim[fil]
```

```python
sig_sim
```

```python
# add a column of the paired DOID to sig_sim
sig_sim['DO1_DO2'] = sig_sim['DO1'] + '_' +sig_sim['DO2'] 
```

```python
sig_sim
```

```python
## add a column of the paired DOID to ont_sim
ont_sim['DO1_DO2'] = ont_sim['DO1'] + '_' + ont_sim['DO2']
```

```python
ont_sim
```

```python
# merge both sig_sim and ont_sim tables together
all_sim = pd.merge(left=sig_sim, right=ont_sim, left_on='DO1_DO2', right_on='DO1_DO2')
```

```python
all_sim
```

```python
# remove the complete correlation (similarity) in case of same ids
fil = all_sim['DZ1'] != all_sim['DZ2']
all_sim = all_sim[fil]
```

```python
all_sim
```

### statistics and plots

```python
standard_deviation= all_sim.SigSim.values.std()
standard_deviation
```

```python
mean = np.mean(all_sim.SigSim.values)
mean
```

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.SigSim.values], color = 'green' , kde = False  )
plt.title('Signature Similarity', fontsize=14)
plt.xlabel('Signature similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
```

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.OntSim.values], color = 'green' , kde = False  )
plt.title('Ontology similarity', fontsize=14)
plt.xlabel('Ontology similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
```

```python
sns.set(color_codes=True)
sns.set(style="whitegrid")
sns.scatterplot( x=all_sim['SigSim'], y=all_sim['OntSim'], marker= '.', color = 'green').set_title('Signature-Ontology Similarity Relationship')
plt.legend(['Relationship'],loc='lower right')
plt.xlabel('Signature-based Jaccard similarity')
plt.ylabel('doSim ontology-based similarity')
```
