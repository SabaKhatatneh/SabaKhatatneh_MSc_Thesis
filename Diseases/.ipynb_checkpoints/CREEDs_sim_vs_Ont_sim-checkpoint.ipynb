```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

### Processing the dataframes 

```python
#open the doSim similarity file
ont_sim = pd.read_csv('~/Downloads/TRP/DrugRepurposing/Diseases/Similarity Tables/doidstable.csv'
                      , sep=',', header=0, index_col=None)
```

```python
# open CREEDS dataset file
meta = pd.read_csv('~/Downloads/TRP/DrugRepurposing/Diseases/disease_signatures-v1.0.csv'
                   , sep=',', header=0, index_col=None)
```

```python
# open CREEDS correlation table file
CREED_sim = pd.read_csv('C:/Users/Saba/Desktop/Tutored research project/DrugRepurposing/data/Diseases/CREEDS similarity.csv'
                      , sep=',', header=0, index_col=None)
CREED_sim
```

```python
# remove self-correlation 
fil = CREED_sim != 1
CREED_sim  = CREED_sim [fil]
```

```python
#reshape the dataframe
CREED_sim = pd.melt(CREED_sim, id_vars='Unnamed: 0', value_vars = CREED_sim.columns[1:])
CREED_sim.columns = ['Dz1', 'Dz2', 'SigSim']
```

```python
CREED_sim
```

```python
# remove null data
fill = CREED_sim.isna().sum(1) == 0
CREED_sim = CREED_sim[fill]
```

```python
CREED_sim
```

```python
# modify the disease id 
CREED_sim.Dz1 = [s.replace("-", ":") for s in CREED_sim.Dz1]
CREED_sim.Dz2 = [s.replace("-", ":") for s in CREED_sim.Dz2]
```

```python
CREED_sim
```

```python
#CREEDS dataset 
meta
```

```python
meta.index = meta['id']
meta = meta['do_id']
```

```python
# adding DOID from meta table to CREED_sim table corresponding disease id
CREED_sim['DO1'] = meta[CREED_sim['Dz1'].values].values
CREED_sim['DO2'] = meta[CREED_sim['Dz2'].values].values
```

```python
CREED_sim
```

```python
# add a column of the paired DOIDs
CREED_sim['DO1_DO2'] = CREED_sim['DO1'] + '_' +CREED_sim['DO2'] 
```

```python
#reshape ont_sim dataframe
ont_sim = pd.melt(ont_sim, id_vars='Unnamed: 0', value_vars=ont_sim.columns[1:])
ont_sim.columns = ['DO1', 'DO2', 'OntSim']
```

```python
#remove complete correlation (self correlation)
fil = ont_sim != 1
ont_sim  = ont_sim [fil]
```

```python
#remove null data
fil = ont_sim.isna().sum(1) == 0
ont_sim = ont_sim[fil]
```

```python
ont_sim
```

```python
# adding the DOID pairs column
ont_sim['DO1_DO2'] = ont_sim['DO1'] + '_' + ont_sim['DO2']
```

```python
# merge both tables together (ont_sim and CREED_sim)
all_sim = pd.merge(left=CREED_sim, right=ont_sim, left_on='DO1_DO2', right_on='DO1_DO2')
```

```python
all_sim
```

### Statistics and plots

```python
sns.set(color_codes=True)
sns.set(style="whitegrid")
sns.scatterplot( x=all_sim['SigSim'], y=all_sim['OntSim'],marker = '.', color = 'green').set_title('CREEDS-Ontology Similarity Relationship')
plt.legend(['Relationship'],loc='lower right')
plt.xlabel('CREEDS Signature correlation')
plt.ylabel('doSim ontology-based similarity')
```

```python
correlation = all_sim.SigSim.corr(all_sim.OntSim)
correlation 
```

```python
pearsonr(all_sim.SigSim, all_sim.OntSim)
```

```python
spearmanr(all_sim.SigSim,all_sim.OntSim)
```
