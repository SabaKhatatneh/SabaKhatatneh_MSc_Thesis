```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

### Open the similarity tables 

```python
up_up = pd.read_csv('../Results/Disease results/up_down genes similarity tables/similarity_up_up.csv'
                      , sep=',', header=0, index_col=None)
up_up
```

```python
dn_dn = pd.read_csv('../Results/Disease results/up_down genes similarity tables/similarity_dn_dn.csv'
                      , sep=',', header=0, index_col=None)
dn_dn
```

```python
dn_up= pd.read_csv('../Results/Disease results/Disease results/up_down genes similarity tables/similarity_dn_up.csv'
                      , sep=',', header=0, index_col=None)
dn_up
```

```python
up_dn= pd.read_csv('../Results/Disease results/Disease results/up_down genes similarity tables/similarity_up_dn.csv'
                      , sep=',', header=0, index_col=None)
up_dn
```

### Reshape the tables 

```python
up_up_sim = pd.melt(up_up, id_vars='Unnamed: 0', value_vars=up_up.columns[1:])
up_up_sim.columns = ['DZ1', 'DZ2', 'UpUp']
```

```python
up_up_sim
```

```python
dn_dn_sim = pd.melt(dn_dn, id_vars='Unnamed: 0', value_vars=dn_dn.columns[1:])
dn_dn_sim.columns = ['DZ1', 'DZ2', 'DnDn']
```

```python
dn_dn_sim
```

```python
dn_up_sim = pd.melt(dn_up, id_vars='Unnamed: 0', value_vars=dn_up.columns[1:])
dn_up_sim.columns = ['DZ1', 'DZ2', 'DnUp']
```

```python
dn_up_sim
```

```python
up_dn_sim = pd.melt(up_dn, id_vars='Unnamed: 0', value_vars=up_dn.columns[1:])
up_dn_sim.columns = ['DZ1', 'DZ2', 'UpDn']
```

```python
up_dn_sim
```

### Merge all tables in one

```python
all_simA= pd.merge(left=up_up_sim, right=dn_dn_sim)
```

```python
all_simA
```

```python
all_simB = pd.merge(left=up_dn_sim, right=dn_up_sim)
```

```python
all_simB
```

```python
all_sim = pd.merge(left=all_simA, right=all_simB)
```

```python
all_sim
```

### plotting histograms for each similarity state

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.UpDn.values], color = 'green' , kde = False  )
plt.title('Up_Down regulated similarity', fontsize=14)
plt.xlabel('Up_Down similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
#,all_sim.DnUp.values, all_sim.UpUp.values, all_sim.DnDn.values
```

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.DnUp.values], color = 'green' , kde = False  )
plt.title('Down_Up regulated similarity', fontsize=14)
plt.xlabel('Down_Up similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
```

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.DnDn.values], color = 'green' , kde = False  )
plt.title('Down-Down regulated similarity', fontsize=14)
plt.xlabel('Down_Down similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
```

```python
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([all_sim.UpUp.values], color = 'green' , kde = False  )
plt.title('Up_Up regulated similarity', fontsize=14)
plt.xlabel('Up_Up similarity', fontsize=11)
plt.ylabel('Density', fontsize=11)
```

```python
sns.set(color_codes=True)
sns.set(style="whitegrid")
sns.scatterplot( x=all_sim['UpUp'], y=all_sim['DnDn'], marker= '.', color = 'green').set_title('Up-Up vs Dn-Dn Similarity Relationship')
plt.legend(['Relationship'],loc='lower right')
plt.xlabel('Up_Up regulated similarity')
plt.ylabel('Dn-Dn regulated similarity')
```
