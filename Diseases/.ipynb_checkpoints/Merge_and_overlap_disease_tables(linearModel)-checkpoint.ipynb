```python
file_name = os.listdir('C:/Users/Saba/Downloads/TRP/DrugRepurposing/Diseases/dz csv')
```

```python
files = [s.replace(".csv", "") for s in file_name]
```

```python
all_genes  = set([])
for file in files:
    data = pd.read_csv('C:/Users/Saba/Downloads/TRP/DrugRepurposing/Diseases/dz csv/%s.csv' % file,
                       sep=',', header=0, index_col=None)
    data['Unnamed: 0'] = data['Unnamed: 0'].str.upper()
    all_genes = all_genes | set(data['Unnamed: 0'])
    data.index= data['Unnamed: 0']
    data = data['logFC']
    files_names = [s.replace("-", ":") for s in files]
    results = pd.DataFrame(index=all_genes, columns=files_names)
    results.loc[data.index, files_names] = data
```

### Compare our results with CREEDS

```python
fin = open('C:/Users/Saba/Downloads/TRP/DrugRepurposing/Diseases/disease_signatures-v1.0.gmt')

flines = fin.readlines()

fin.close()

#print (len(flines))

flines[2].split('\t')

#print (flines[0])

```

```python
diseses_genes = {}
for i in range(len(flines)):
    actual_disese = flines[i][:-1].split('\t')
    diseaseId = actual_disese[1]
    direction = actual_disese[0].split('-')[-1]
    genes = actual_disese[2:]
    genes = [x.upper() for x in genes]
        
    
    if diseaseId not in diseses_genes.keys():
        diseses_genes[diseaseId] = {} 
    diseses_genes[diseaseId][direction] = genes
```

```python
### get the same genes from creeds and results table (GEO-linear model)
up = diseses_genes['dz:1000']['up']
dn = diseses_genes['dz:1000']['dn']
up = list(set(results.index) & set(up))
dn = list(set(results.index) & set(dn))
```

```python
plt.hist(results['dz:1000'], label = 'All', density=True)
plt.xlabel('dz:1000')
plt.ylabel('number of genes')
plt.title('dz:1000')
plt.hist(results['dz:1000'][up], density=True , label = 'Up')
plt.hist(results['dz:1000'][dn], density=True, label = 'Down')
plt.legend(loc='upper right')
```
