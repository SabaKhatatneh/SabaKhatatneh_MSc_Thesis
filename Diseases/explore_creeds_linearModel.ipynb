```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('../Results/Disease results//merge_and_overlap.csv', 
                   sep=',', header=0, index_col=0)
data
```

```python
#number of nans for each disease
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(0)], color = 'green' , kde = False  )
plt.title('Null Values for each disease', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Diseases', fontsize=11)
```

```python
#number of nans for each gene
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(1)], color = 'green' , kde = False  )
plt.title('Null Values for each gene', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Genes', fontsize=11)
```

```python
### remove genes that are null in most of diseases
fil = data.isna().sum(1) < 700
data = data[fil]
```

```python
#number of nans for each disease <700
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(0)], color = 'green' , kde = False  )
plt.title('Null Values for each disease', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Diseases', fontsize=11)
```

```python
#number of nans for each gene <700
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(1)], color = 'green' , kde = False  )
plt.title('Null Values for each gene', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Genes', fontsize=11)
```

```python
### remove diseases that are mostly null
fil = data.isna().sum(0) < 17500
data = data[data.columns[fil]]
```

```python
data
```

```python
#number of nans for each disease < 17500
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(0)], color = 'green' , kde = False  )
plt.title('Null Values for each disease', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Diseases', fontsize=11)
```

```python
#number of nans for each gene < 17500
sns.set_theme()
sns.set(style="darkgrid")
sns.distplot([data.isna().sum(1)], color = 'green' , kde = False  )
plt.title('Null Values for each gene', fontsize=14)
plt.xlabel('Null values', fontsize=11)
plt.ylabel('Genes', fontsize=11)
```

```python
#Check a random disease Id pairs relationship
sns.set(style="whitegrid")
sns.scatterplot( x=data['dz-1000'], y=data['dz-1001'], color = 'green').set_title('dz-1000 vs dz-1001 Relationship')
plt.legend(['Relationship'],loc='lower right')
plt.xlabel('dz-1000')
plt.ylabel('dz-1001')
data['dz-1000'].corr(data['dz-1001'])
```

```python
#Check a random disease Id pairs relationship
sns.set(style="whitegrid")
sns.scatterplot( x=data['dz-1000'], y=data['dz-323'], color = 'green').set_title('dz-1000 vs dz-323 Relationship')
plt.legend(['Relationship'],loc='lower right')
plt.xlabel('dz-1000')
plt.ylabel('dz-323')
data['dz-1000'].corr(data['dz-323'])
```
